```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make2d(N,r=1,noise=0.0001):
    X = []
    c = []

    for _ in range(N):
        while True:
            x = np.random.uniform(-r, r)
            y = np.random.uniform(-r, r)

            if x**2 + y**2 <= r**2:
                if y >= 0: c_ = 0
                else: c_ = 1

                if y >= 0 and (x + r/2)**2 + y**2 <= (r/2)**2:
                    c_ = 1 - c_
                elif y < 0 and (x - r/2)**2 + y**2 <= (r/2)**2:
                    c_ = 1 - c_

                x += np.random.normal(0, noise)
                y += np.random.normal(0, noise)

                X.append([x, y])
                c.append(c_)
                break

    X,c = np.array(X), np.array(c)

    return X, c


N=200
r=1
noise= 0.1
X,c= make2d(N,r,noise)


d = 1.5
X[c==1,1]-=d


plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=c, cmap='bwr', alpha=0.7)
plt.show()
```
